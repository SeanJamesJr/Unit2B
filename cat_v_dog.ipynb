```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as img
import numpy as np
traning_data = os.listdir("train/")[:14999]
print(traning_data)
```

```python
labels =[]

for x in traning_data:
    if 'cat' in x:
        labels.append(0)
    else:
        labels.append(1)
       
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as img

cat_image = img.imread("train/cat.10300.jpg")
plt.imshow(cat_image)
```

```python
import PIL

for no in traning_data:
 img = PIL.Image.open('train/'+ no)
 img= img.resize((224,224))
 img =img.convert('RGB')

 img.save('resized/' + no)
```

```python
import cv2
import numpy as np

saved_img =[]
for yes in traning_data:
    img = cv2.imread('resized/'+ yes)
    saved_img.append(img)

saved_img = np.asarray(saved_img)
```

```python
from sklearn.model_selection import train_test_split

labels = np.asarray(labels)

x_train, x_test, y_train, y_test=train_test_split(saved_img, labels,test_size=0.2,random_state=2)
x_train_scaled = x_train/255
x_test_scaled =x_test/255
```

```python
import tensorflow as tf
import tensorflow_hub as hub


mobilenet_model = 'https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4'

pretrained_model = hub.KerasLayer(mobilenet_model, input_shape=(224,224,3), trainable=False)
```

```python
model = tf.keras.Sequential([
    pretrained_model,
    tf.keras.layers.Dense(2)
])

model.summary()
```

```python
model.compile(
    optimizer = 'adam',
    loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics = ['acc']
)
```
